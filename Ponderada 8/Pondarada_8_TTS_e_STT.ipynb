```python
import whisper

def transcribe_audio(file_path):
    # Carregar o modelo Whisper
    model = whisper.load_model("medium")

    # Carregar o arquivo de áudio
    audio = whisper.load_audio(file_path)

    # Transcrever o áudio diretamente para inglês
    result = model.transcribe(audio, language='en')

    # Extrair e retornar o texto transcrito
    text = result['text']
    return text
```

```python
from bark import SAMPLE_RATE, generate_audio, preload_models

def synthesize_audio(text):
    preload_models()
    audio_array = generate_audio(text)
    return audio_array
```

```python
def save_text_to_file(text, file_name):
    with open(file_name, 'w', encoding='utf-8') as file:
        file.write(text)
        print(f"Texto gerado salvo!")
```

```python
from scipy.io.wavfile import write as write_wav

def save_speech_to_file(audio_array):
    write_wav("./Midia/Gerados/bark_generation_calma.wav", SAMPLE_RATE, audio_array)
    print(f"Áudio gerado salvo!")
    return
```

```python
def main():
    # Obter caminho do arquivo de áudio
    # audio_path = "./omae.mp3"

    audio_path = input("Enter the path to the audio file: ")

    # Transcrever áudio para texto
    transcribed_text = transcribe_audio(audio_path)
    print("Texto Transcrito:", transcribed_text)

    save_text_to_file(transcribed_text, "./Midia/Gerados/transcribed_text.txt")  # Salvando o texto transcrita

    # Sintetizar áudio da tradução
    audio = synthesize_audio(transcribed_text)

    save_speech_to_file(audio)

if __name__ == "__main__":
    main()
```
